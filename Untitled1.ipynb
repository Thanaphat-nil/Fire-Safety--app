```python
import os
import sqlite3
from datetime import date, datetime, timedelta
from typing import Optional, Tuple

import pandas as pd
import streamlit as st

DB_PATH = "fire_extinguishers.db"
DATE_FMT = "%Y-%m-%d"

# ----------------------------- DB LAYER -----------------------------

def get_conn():
    return sqlite3.connect(DB_PATH, check_same_thread=False)

def init_db():
    with get_conn() as conn:
        cur = conn.cursor()
        # Extinguishers master table
        cur.execute(
            """
            CREATE TABLE IF NOT EXISTS extinguishers (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                serial TEXT UNIQUE NOT NULL,
                location TEXT NOT NULL,
                type TEXT NOT NULL,
                capacity_kg REAL,
                manufacturer TEXT,
                manufacture_date DATE,
                in_service_date DATE,
                active INTEGER DEFAULT 1,
                notes TEXT
            );
            """
        )
        # Inspections table
        cur.execute(
            """
            CREATE TABLE IF NOT EXISTS inspections (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                extinguisher_id INTEGER NOT NULL,
                inspection_date DATE NOT NULL,
                inspector TEXT,
                gauge_ok INTEGER,
                seal_ok INTEGER,
                pin_ok INTEGER,
                hose_nozzle_ok INTEGER,
                clear_access INTEGER,
                notes TEXT,
                FOREIGN KEY (extinguisher_id) REFERENCES extinguishers (id)
            );
            """
        )
        conn.commit()

def add_extinguisher(
    serial: str,
    location: str,
    type: str,
    capacity_kg: Optional[float] = None,
    manufacturer: Optional[str] = None,
    manufacture_date: Optional[date] = None,
    in_service_date: Optional[date] = None,
    notes: Optional[str] = None,
):
    with get_conn() as conn:
        cur = conn.cursor()
        cur.execute(
            """
            INSERT INTO extinguishers (
                serial,
                location,
                type,
                capacity_kg,
                manufacturer,
                manufacture_date,
                in_service_date,
                notes
            ) VALUES (?, ?, ?, ?, ?, ?, ?, ?);
            """,
            (
                serial,
                location,
                type,
                capacity_kg,
                manufacturer,
                manufacture_date.strftime(DATE_FMT) if manufacture_date else None,
                in_service_date.strftime(DATE_FMT) if in_service_date else None,
                notes,
            ),
        )
        conn.commit()

def get_extinguishers(active_only: bool = True):
    with get_conn() as conn:
        cur = conn.cursor()
        query = "SELECT * FROM extinguishers"
        if active_only:
            query += " WHERE active = 1"
        cur.execute(query)
        rows = cur.fetchall()
        columns = [description[0] for description in cur.description]
        return pd.DataFrame(rows, columns=columns)

def update_extinguisher(extinguisher_id, **kwargs):
    with get_conn() as conn:
        cur = conn.cursor()
        set_clauses = [f"{key} = ?" for key in kwargs.keys()]
        query = f"UPDATE extinguishers SET {', '.join(set_clauses)} WHERE id = ?"
        values = list(kwargs.values())
        values.append(extinguisher_id)
        cur.execute(query, values)
        conn.commit()

def add_inspection(
    extinguisher_id: int,
    inspection_date: date,
    inspector: Optional[str] = None,
    gauge_ok: Optional[int] = None,
    seal_ok: Optional[int] = None,
    pin_ok: Optional[int] = None,
    hose_nozzle_ok: Optional[int] = None,
    clear_access: Optional[int] = None,
    notes: Optional[str] = None,
):
    with get_conn() as conn:
        cur = conn.cursor()
        cur.execute(
            """
            INSERT INTO inspections (
                extinguisher_id,
                inspection_date,
                inspector,
                gauge_ok,
                seal_ok,
                pin_ok,
                hose_nozzle_ok,
                clear_access,
                notes
            ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?);
            """,
            (
                extinguisher_id,
                inspection_date.strftime(DATE_FMT),
                inspector,
                gauge_ok,
                seal_ok,
                pin_ok,
                hose_nozzle_ok,
                clear_access,
                notes,
            ),
        )
        conn.commit()

def get_inspections(extinguisher_id: Optional[int] = None):
    with get_conn() as conn:
        cur = conn.cursor()
        query = "SELECT * FROM inspections"
        if extinguisher_id:
            query += f" WHERE extinguisher_id = {extinguisher_id}"
        cur.execute(query)
        rows = cur.fetchall()
        columns = [description[0] for description in cur.description]
        return pd.DataFrame(rows, columns=columns)

def get_last_inspection_date(extinguisher_id: int) -> Optional[date]:
    with get_conn() as conn:
        cur = conn.cursor()
        cur.execute(
            "SELECT MAX(inspection_date) FROM inspections WHERE extinguisher_id = ?",
            (extinguisher_id,),
        )
        result = cur.fetchone()[0]
        return datetime.strptime(result, DATE_FMT).date() if result else None

def import_extinguishers_from_csv(filepath: str):
    df = pd.read_csv(filepath)
    df = df.rename(columns={
        'Serial Number': 'serial',
        'Location': 'location',
        'Type': 'type',
        'Capacity (kg)': 'capacity_kg',
        'Manufacturer': 'manufacturer',
        'Manufacture Date': 'manufacture_date',
        'In Service Date': 'in_service_date',
        'Notes': 'notes'
    })
    df['manufacture_date'] = pd.to_datetime(df['manufacture_date'], errors='coerce').dt.date
    df['in_service_date'] = pd.to_datetime(df['in_service_date'], errors='coerce').dt.date

    with get_conn() as conn:
        cur = conn.cursor()
        for index, row in df.iterrows():
            try:
                cur.execute(
                    """
                    INSERT INTO extinguishers (
                        serial,
                        location,
                        type,
                        capacity_kg,
                        manufacturer,
                        manufacture_date,
                        in_service_date,
                        notes
                    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?)
                    ON CONFLICT(serial) DO UPDATE SET
                        location = excluded.location,
                        type = excluded.type,
                        capacity_kg = excluded.capacity_kg,
                        manufacturer = excluded.manufacturer,
                        manufacture_date = excluded.manufacture_date,
                        in_service_date = excluded.in_service_date,
                        notes = excluded.notes;
                    """,
                    (
                        row['serial'],
                        row['location'],
                        row['type'],
                        row['capacity_kg'],
                        row['manufacturer'],
                        row['manufacture_date'].strftime(DATE_FMT) if pd.notna(row['manufacture_date']) else None,
                        row['in_service_date'].strftime(DATE_FMT) if pd.notna(row['in_service_date']) else None,
                        row['notes'],
                    ),
                )
            except Exception as e:
                st.error(f"Error importing row {index}: {e}")
        conn.commit()


# ----------------------------- APP UI -----------------------------

st.set_page_config(layout="wide")

def main():
    init_db()

    menu = ["Extinguishers", "Inspections", "Due/Overdue", "Reports", "Settings"]
    choice = st.sidebar.selectbox("Menu", menu)

    if choice == "Extinguishers":
        manage_extinguishers()
    elif choice == "Inspections":
        log_inspection()
    elif choice == "Due/Overdue":
        view_due_overdue()
    elif choice == "Reports":
        generate_reports()
    elif choice == "Settings":
        app_settings()

def manage_extinguishers():
    st.subheader("Manage Fire Extinguishers")

    # Add new extinguisher
    with st.expander("Add New Extinguisher"):
        with st.form("add_extinguisher_form"):
            serial = st.text_input("Serial Number", key="add_serial")
            location = st.text_input("Location", key="add_location")
            type = st.text_input("Type", key="add_type")
            capacity_kg = st.number_input("Capacity (kg)", key="add_capacity", min_value=0.0, format="%.2f")
            manufacturer = st.text_input("Manufacturer", key="add_manufacturer")
            manufacture_date = st.date_input("Manufacture Date", key="add_manufacture_date", value=None)
            in_service_date = st.date_input("In Service Date", key="add_in_service_date", value=None)
            notes = st.text_area("Notes", key="add_notes")

            submitted = st.form_submit_button("Add Extinguisher")
            if submitted:
                if not serial or not location or not type:
                    st.error("Serial Number, Location, and Type are required.")
                else:
                    add_extinguisher(
                        serial,
                        location,
                        type,
                        capacity_kg if capacity_kg > 0 else None,
                        manufacturer,
                        manufacture_date,
                        in_service_date,
                        notes,
                    )
                    st.success(f"Extinguisher {serial} added.")

    # View/Edit extinguishers
    st.subheader("Existing Extinguishers")
    extinguishers_df = get_extinguishers(active_only=False)
    if not extinguishers_df.empty:
        st.dataframe(extinguishers_df)

        with st.expander("Edit Extinguisher"):
            extinguisher_id_edit = st.number_input("Enter Extinguisher ID to Edit", min_value=1)
            edit_extinguisher_df = extinguishers_df[extinguishers_df['id'] == extinguisher_id_edit]
            if not edit_extinguisher_df.empty:
                extinguisher_data = edit_extinguisher_df.iloc[0]
                with st.form("edit_extinguisher_form"):
                    serial_edit = st.text_input("Serial Number", value=extinguisher_data['serial'], key="edit_serial")
                    location_edit = st.text_input("Location", value=extinguisher_data['location'], key="edit_location")
                    type_edit = st.text_input("Type", value=extinguisher_data['type'], key="edit_type")
                    capacity_kg_edit = st.number_input("Capacity (kg)", value=extinguisher_data['capacity_kg'] if pd.notna(extinguisher_data['capacity_kg']) else 0.0, min_value=0.0, format="%.2f", key="edit_capacity")
                    manufacturer_edit = st.text_input("Manufacturer", value=extinguisher_data['manufacturer'] if pd.notna(extinguisher_data['manufacturer']) else "", key="edit_manufacturer")
                    manufacture_date_edit = st.date_input("Manufacture Date", value=datetime.strptime(extinguisher_data['manufacture_date'], DATE_FMT).date() if pd.notna(extinguisher_data['manufacture_date']) else None, key="edit_manufacture_date")
                    in_service_date_edit = st.date_input("In Service Date", value=datetime.strptime(extinguisher_data['in_service_date'], DATE_FMT).date() if pd.notna(extinguisher_data['in_service_date']) else None, key="edit_in_service_date")
                    active_edit = st.checkbox("Active", value=bool(extinguisher_data['active']), key="edit_active")
                    notes_edit = st.text_area("Notes", value=extinguisher_data['notes'] if pd.notna(extinguisher_data['notes']) else "", key="edit_notes")

                    submitted_edit = st.form_submit_button("Update Extinguisher")
                    if submitted_edit:
                        update_extinguisher(
                            extinguisher_id_edit,
                            serial=serial_edit,
                            location=location_edit,
                            type=type_edit,
                            capacity_kg=capacity_kg_edit if capacity_kg_edit > 0 else None,
                            manufacturer=manufacturer_edit,
                            manufacture_date=manufacture_date_edit,
                            in_service_date=in_service_date_edit,
                            active=int(active_edit),
                            notes=notes_edit,
                        )
                        st.success(f"Extinguisher {serial_edit} updated.")
                        st.experimental_rerun()
            elif extinguisher_id_edit != 1:
                 st.warning("Extinguisher ID not found.")

    else:
        st.info("No extinguishers registered yet.")


def log_inspection():
    st.subheader("Log Monthly Inspection")
    extinguishers_df = get_extinguishers()
    if extinguishers_df.empty:
        st.warning("No active extinguishers to inspect. Please add extinguishers first.")
        return

    extinguisher_serials = extinguishers_df['serial'].tolist()
    selected_serial = st.selectbox("Select Extinguisher", extinguisher_serials)
    selected_extinguisher_id = extinguishers_df[extinguishers_df['serial'] == selected_serial]['id'].iloc[0]

    last_inspection_date = get_last_inspection_date(selected_extinguisher_id)
    if last_inspection_date:
        st.info(f"Last inspection on: {last_inspection_date.strftime(DATE_FMT)}")
    else:
        st.info("No previous inspections recorded for this extinguisher.")

    with st.form("log_inspection_form"):
        inspection_date = st.date_input("Inspection Date", value=date.today())
        inspector = st.text_input("Inspector Name")
        gauge_ok = st.checkbox("Pressure Gauge OK?")
        seal_ok = st.checkbox("Seal Intact?")
        pin_ok = st.checkbox("Safety Pin OK?")
        hose_nozzle_ok = st.checkbox("Hose/Nozzle OK?")
        clear_access = st.checkbox("Clear Access?")
        notes = st.text_area("Notes")

        submitted = st.form_submit_button("Log Inspection")
        if submitted:
            add_inspection(
                selected_extinguisher_id,
                inspection_date,
                inspector,
                int(gauge_ok),
                int(seal_ok),
                int(pin_ok),
                int(hose_nozzle_ok),
                int(clear_access),
                notes,
            )
            st.success(f"Inspection logged for {selected_serial} on {inspection_date.strftime(DATE_FMT)}.")
            st.experimental_rerun()

    st.subheader("Recent Inspections")
    inspections_df = get_inspections(extinguisher_id=selected_extinguisher_id)
    if not inspections_df.empty:
        st.dataframe(inspections_df)
    else:
        st.info("No inspections recorded for this extinguisher.")


def view_due_overdue():
    st.subheader("Due/Overdue Inspections")

    extinguishers_df = get_extinguishers()
    if extinguishers_df.empty:
        st.info("No active extinguishers to check.")
        return

    today = date.today()
    due_inspections = []

    for index, row in extinguishers_df.iterrows():
        extinguisher_id = row['id']
        last_inspection_date = get_last_inspection_date(extinguisher_id)

        if last_inspection_date:
            next_due_date = last_inspection_date + timedelta(days=30) # Assuming monthly inspection
            status = "Due" if next_due_date <= today else "OK"
            if status == "Due":
                due_inspections.append({
                    "Serial Number": row['serial'],
                    "Location": row['location'],
                    "Last Inspection Date": last_inspection_date.strftime(DATE_FMT),
                    "Next Due Date": next_due_date.strftime(DATE_FMT),
                    "Status": "Overdue" if next_due_date < today else "Due"
                })
        else:
             # Extinguisher never inspected
             due_inspections.append({
                "Serial Number": row['serial'],
                "Location": row['location'],
                "Last Inspection Date": "N/A",
                "Next Due Date": "Due Now",
                "Status": "Due"
            })

    due_df = pd.DataFrame(due_inspections)
    if not due_df.empty:
        st.dataframe(due_df)
    else:
        st.info("All active extinguishers are up-to-date with inspections.")


def generate_reports():
    st.subheader("Monthly Inspection Reports")

    extinguishers_df = get_extinguishers(active_only=False)
    inspections_df = get_inspections()

    if inspections_df.empty:
        st.info("No inspections recorded to generate reports.")
        return

    inspections_df['inspection_date'] = pd.to_datetime(inspections_df['inspection_date'])
    inspections_df['month_year'] = inspections_df['inspection_date'].dt.to_period('M')

    # Merge with extinguishers info
    report_df = pd.merge(inspections_df, extinguishers_df, left_on='extinguisher_id', right_on='id', suffixes=('_inspection', '_extinguisher'))

    # Select and rename columns for the report
    report_df = report_df[[
        'month_year',
        'serial',
        'location',
        'type',
        'inspection_date',
        'inspector',
        'gauge_ok',
        'seal_ok',
        'pin_ok',
        'hose_nozzle_ok',
        'clear_access',
        'notes_inspection'
    ]]
    report_df = report_df.rename(columns={'notes_inspection': 'inspection_notes'})


    st.dataframe(report_df)

    # Export as CSV
    csv_data = report_df.to_csv(index=False).encode('utf-8')
    st.download_button(
        label="Download Report as CSV",
        data=csv_data,
        file_name="fire_extinguisher_inspection_report.csv",
        mime="text/csv",
    )


def app_settings():
    st.subheader("Settings")

    st.write("### Import Extinguishers from CSV")
    st.write("Upload a CSV file with columns: 'Serial Number', 'Location', 'Type', 'Capacity (kg)', 'Manufacturer', 'Manufacture Date', 'In Service Date', 'Notes'")
    uploaded_file = st.file_uploader("Choose a CSV file", type="csv")

    if uploaded_file is not None:
        try:
            # Save the uploaded file temporarily
            filepath = os.path.join("/tmp", uploaded_file.name)
            with open(filepath, "wb") as f:
                f.write(uploaded_file.getbuffer())

            import_extinguishers_from_csv(filepath)
            st.success("Extinguishers imported successfully.")
            os.remove(filepath) # Clean up the temporary file
        except Exception as e:
            st.error(f"Error importing CSV: {e}")

if __name__ == "__main__":
    main()
```

```python
%%writefile fire_extinguisher_inspection_app.py
import os
import sqlite3
from datetime import date, datetime, timedelta
from typing import Optional, Tuple

import pandas as pd
import streamlit as st

DB_PATH = "fire_extinguishers.db"
DATE_FMT = "%Y-%m-%d"

# ----------------------------- DB LAYER -----------------------------

def get_conn():
    return sqlite3.connect(DB_PATH, check_same_thread=False)

def init_db():
    with get_conn() as conn:
        cur = conn.cursor()
        # Extinguishers master table
        cur.execute(
            """
            CREATE TABLE IF NOT EXISTS extinguishers (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                serial TEXT UNIQUE NOT NULL,
                location TEXT NOT NULL,
                type TEXT NOT NULL,
                capacity_kg REAL,
                manufacturer TEXT,
                manufacture_date DATE,
                in_service_date DATE,
                active INTEGER DEFAULT 1,
                notes TEXT
            );
            """
        )
        # Inspections table
        cur.execute(
            """
            CREATE TABLE IF NOT EXISTS inspections (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                extinguisher_id INTEGER NOT NULL,
                inspection_date DATE NOT NULL,
                inspector TEXT,
                gauge_ok INTEGER,
                seal_ok INTEGER,
                pin_ok INTEGER,
                hose_nozzle_ok INTEGER,
                clear_access INTEGER,
                notes TEXT,
                FOREIGN KEY (extinguisher_id) REFERENCES extinguishers (id)
            );
            """
        )
        conn.commit()

def add_extinguisher(
    serial: str,
    location: str,
    type: str,
    capacity_kg: Optional[float] = None,
    manufacturer: Optional[str] = None,
    manufacture_date: Optional[date] = None,
    in_service_date: Optional[date] = None,
    notes: Optional[str] = None,
):
    with get_conn() as conn:
        cur = conn.cursor()
        cur.execute(
            """
            INSERT INTO extinguishers (
                serial,
                location,
                type,
                capacity_kg,
                manufacturer,
                manufacture_date,
                in_service_date,
                notes
            ) VALUES (?, ?, ?, ?, ?, ?, ?, ?);
            """,
            (
                serial,
                location,
                type,
                capacity_kg,
                manufacture_date.strftime(DATE_FMT) if manufacture_date else None,
                in_service_date.strftime(DATE_FMT) if in_service_date else None,
                notes,
            ),
        )
        conn.commit()

def get_extinguishers(active_only: bool = True):
    with get_conn() as conn:
        cur = conn.cursor()
        query = "SELECT * FROM extinguishers"
        if active_only:
            query += " WHERE active = 1"
        cur.execute(query)
        rows = cur.fetchall()
        columns = [description[0] for description in cur.description]
        return pd.DataFrame(rows, columns=columns)

def update_extinguisher(extinguisher_id, **kwargs):
    with get_conn() as conn:
        cur = conn.cursor()
        set_clauses = [f"{key} = ?" for key in kwargs.keys()]
        query = f"UPDATE extinguishers SET {', '.join(set_clauses)} WHERE id = ?"
        values = list(kwargs.values())
        values.append(extinguisher_id)
        cur.execute(query, values)
        conn.commit()

def add_inspection(
    extinguisher_id: int,
    inspection_date: date,
    inspector: Optional[str] = None,
    gauge_ok: Optional[int] = None,
    seal_ok: Optional[int] = None,
    pin_ok: Optional[int] = None,
    hose_nozzle_ok: Optional[int] = None,
    clear_access: Optional[int] = None,
    notes: Optional[str] = None,
):
    with get_conn() as conn:
        cur = conn.cursor()
        cur.execute(
            """
            INSERT INTO inspections (
                extinguisher_id,
                inspection_date,
                inspector,
                gauge_ok,
                seal_ok,
                pin_ok,
                hose_nozzle_ok,
                clear_access,
                notes
            ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?);
            """,
            (
                extinguisher_id,
                inspection_date.strftime(DATE_FMT),
                inspector,
                gauge_ok,
                seal_ok,
                pin_ok,
                hose_nozzle_ok,
                clear_access,
                notes,
            ),
        )
        conn.commit()

def get_inspections(extinguisher_id: Optional[int] = None):
    with get_conn() as conn:
        cur = conn.cursor()
        query = "SELECT * FROM inspections"
        if extinguisher_id:
            query += f" WHERE extinguisher_id = {extinguisher_id}"
        cur.execute(query)
        rows = cur.fetchall()
        columns = [description[0] for description in cur.description]
        return pd.DataFrame(rows, columns=columns)

def get_last_inspection_date(extinguisher_id: int) -> Optional[date]:
    with get_conn() as conn:
        cur = conn.cursor()
        cur.execute(
            "SELECT MAX(inspection_date) FROM inspections WHERE extinguisher_id = ?",
            (extinguisher_id,),
        )
        result = cur.fetchone()[0]
        return datetime.strptime(result, DATE_FMT).date() if result else None

def import_extinguishers_from_csv(filepath: str):
    df = pd.read_csv(filepath)
    df = df.rename(columns={
        'Serial Number': 'serial',
        'Location': 'location',
        'Type': 'type',
        'Capacity (kg)': 'capacity_kg',
        'Manufacturer': 'manufacturer',
        'Manufacture Date': 'manufacture_date',
        'In Service Date': 'in_service_date',
        'Notes': 'notes'
    })
    df['manufacture_date'] = pd.to_datetime(df['manufacture_date'], errors='coerce').dt.date
    df['in_service_date'] = pd.to_datetime(df['in_service_date'], errors='coerce').dt.date

    with get_conn() as conn:
        cur = conn.cursor()
        for index, row in df.iterrows():
            try:
                cur.execute(
                    """
                    INSERT INTO extinguishers (
                        serial,
                        location,
                        type,
                        capacity_kg,
                        manufacturer,
                        manufacture_date,
                        in_service_date,
                        notes
                    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?)
                    ON CONFLICT(serial) DO UPDATE SET
                        location = excluded.location,
                        type = excluded.type,
                        capacity_kg = excluded.capacity_kg,
                        manufacturer = excluded.manufacturer,
                        manufacture_date = excluded.manufacture_date,
                        in_service_date = excluded.in_service_date,
                        notes = excluded.notes;
                    """,
                    (
                        row['serial'],
                        row['location'],
                        row['type'],
                        row['capacity_kg'],
                        row['manufacturer'],
                        row['manufacture_date'].strftime(DATE_FMT) if pd.notna(row['manufacture_date']) else None,
                        row['in_service_date'].strftime(DATE_FMT) if pd.notna(row['in_service_date']) else None,
                        row['notes'],
                    ),
                )
            except Exception as e:
                st.error(f"Error importing row {index}: {e}")
        conn.commit()


# ----------------------------- APP UI -----------------------------

st.set_page_config(layout="wide")

def main():
    init_db()

    menu = ["Extinguishers", "Inspections", "Due/Overdue", "Reports", "Settings"]
    choice = st.sidebar.selectbox("Menu", menu)

    if choice == "Extinguishers":
        manage_extinguishers()
    elif choice == "Inspections":
        log_inspection()
    elif choice == "Due/Overdue":
        view_due_overdue()
    elif choice == "Reports":
        generate_reports()
    elif choice == "Settings":
        app_settings()

def manage_extinguishers():
    st.subheader("Manage Fire Extinguishers")

    # Add new extinguisher
    with st.expander("Add New Extinguisher"):
        with st.form("add_extinguisher_form"):
            serial = st.text_input("Serial Number", key="add_serial")
            location = st.text_input("Location", key="add_location")
            type = st.text_input("Type", key="add_type")
            capacity_kg = st.number_input("Capacity (kg)", key="add_capacity", min_value=0.0, format="%.2f")
            manufacturer = st.text_input("Manufacturer", key="add_manufacturer")
            manufacture_date = st.date_input("Manufacture Date", key="add_manufacture_date", value=None)
            in_service_date = st.date_input("In Service Date", key="add_in_service_date", value=None)
            notes = st.text_area("Notes", key="add_notes")

            submitted = st.form_submit_button("Add Extinguisher")
            if submitted:
                if not serial or not location or not type:
                    st.error("Serial Number, Location, and Type are required.")
                else:
                    add_extinguisher(
                        serial,
                        location,
                        type,
                        capacity_kg if capacity_kg > 0 else None,
                        manufacturer,
                        manufacture_date,
                        in_service_date,
                        notes,
                    )
                    st.success(f"Extinguisher {serial} added.")

    # View/Edit extinguishers
    st.subheader("Existing Extinguishers")
    extinguishers_df = get_extinguishers(active_only=False)
    if not extinguishers_df.empty:
        st.dataframe(extinguishers_df)

        with st.expander("Edit Extinguisher"):
            extinguisher_id_edit = st.number_input("Enter Extinguisher ID to Edit", min_value=1)
            edit_extinguisher_df = extinguishers_df[extinguishers_df['id'] == extinguisher_id_edit]
            if not edit_extinguisher_df.empty:
                extinguisher_data = edit_extinguisher_df.iloc[0]
                with st.form("edit_extinguisher_form"):
                    serial_edit = st.text_input("Serial Number", value=extinguisher_data['serial'], key="edit_serial")
                    location_edit = st.text_input("Location", value=extinguisher_data['location'], key="edit_location")
                    type_edit = st.text_input("Type", value=extinguisher_data['type'], key="edit_type")
                    capacity_kg_edit = st.number_input("Capacity (kg)", value=extinguisher_data['capacity_kg'] if pd.notna(extinguisher_data['capacity_kg']) else 0.0, min_value=0.0, format="%.2f", key="edit_capacity")
                    manufacturer_edit = st.text_input("Manufacturer", value=extinguisher_data['manufacturer'] if pd.notna(extinguisher_data['manufacturer']) else "", key="edit_manufacturer")
                    manufacture_date_edit = st.date_input("Manufacture Date", value=datetime.strptime(extinguisher_data['manufacture_date'], DATE_FMT).date() if pd.notna(extinguisher_data['manufacture_date']) else None, key="edit_manufacture_date")
                    in_service_date_edit = st.date_input("In Service Date", value=datetime.strptime(extinguisher_data['in_service_date'], DATE_FMT).date() if pd.notna(extinguisher_data['in_service_date']) else None, key="edit_in_service_date")
                    active_edit = st.checkbox("Active", value=bool(extinguisher_data['active']), key="edit_active")
                    notes_edit = st.text_area("Notes", value=extinguisher_data['notes'] if pd.notna(extinguisher_data['notes']) else "", key="edit_notes")

                    submitted_edit = st.form_submit_button("Update Extinguisher")
                    if submitted_edit:
                        update_extinguisher(
                            extinguisher_id_edit,
                            serial=serial_edit,
                            location=location_edit,
                            type=type_edit,
                            capacity_kg=capacity_kg_edit if capacity_kg_edit > 0 else None,
                            manufacturer=manufacturer_edit,
                            manufacture_date=manufacture_date_edit,
                            in_service_date=in_service_date_edit,
                            active=int(active_edit),
                            notes=notes_edit,
                        )
                        st.success(f"Extinguisher {serial_edit} updated.")
                        st.experimental_rerun()
            elif extinguisher_id_edit != 1:
                 st.warning("Extinguisher ID not found.")

    else:
        st.info("No extinguishers registered yet.")


def log_inspection():
    st.subheader("Log Monthly Inspection")
    extinguishers_df = get_extinguishers()
    if extinguishers_df.empty:
        st.warning("No active extinguishers to inspect. Please add extinguishers first.")
        return

    extinguisher_serials = extinguishers_df['serial'].tolist()
    selected_serial = st.selectbox("Select Extinguisher", extinguisher_serials)
    selected_extinguisher_id = extinguishers_df[extinguishers_df['serial'] == selected_serial]['id'].iloc[0]

    last_inspection_date = get_last_inspection_date(selected_extinguisher_id)
    if last_inspection_date:
        st.info(f"Last inspection on: {last_inspection_date.strftime(DATE_FMT)}")
    else:
        st.info("No previous inspections recorded for this extinguisher.")

    with st.form("log_inspection_form"):
        inspection_date = st.date_input("Inspection Date", value=date.today())
        inspector = st.text_input("Inspector Name")
        gauge_ok = st.checkbox("Pressure Gauge OK?")
        seal_ok = st.checkbox("Seal Intact?")
        pin_ok = st.checkbox("Safety Pin OK?")
        hose_nozzle_ok = st.checkbox("Hose/Nozzle OK?")
        clear_access = st.checkbox("Clear Access?")
        notes = st.text_area("Notes")

        submitted = st.form_submit_button("Log Inspection")
        if submitted:
            add_inspection(
                selected_extinguisher_id,
                inspection_date,
                inspector,
                int(gauge_ok),
                int(seal_ok),
                int(pin_ok),
                int(hose_nozzle_ok),
                int(clear_access),
                notes,
            )
            st.success(f"Inspection logged for {selected_serial} on {inspection_date.strftime(DATE_FMT)}.")
            st.experimental_rerun()

    st.subheader("Recent Inspections")
    inspections_df = get_inspections(extinguisher_id=selected_extinguisher_id)
    if not inspections_df.empty:
        st.dataframe(inspections_df)
    else:
        st.info("No inspections recorded for this extinguisher.")


def view_due_overdue():
    st.subheader("Due/Overdue Inspections")

    extinguishers_df = get_extinguishers()
    if extinguishers_df.empty:
        st.info("No active extinguishers to check.")
        return

    today = date.today()
    due_inspections = []

    for index, row in extinguishers_df.iterrows():
        extinguisher_id = row['id']
        last_inspection_date = get_last_inspection_date(extinguisher_id)

        if last_inspection_date:
            next_due_date = last_inspection_date + timedelta(days=30) # Assuming monthly inspection
            status = "Due" if next_due_date <= today else "OK"
            if status == "Due":
                due_inspections.append({
                    "Serial Number": row['serial'],
                    "Location": row['location'],
                    "Last Inspection Date": last_inspection_date.strftime(DATE_FMT),
                    "Next Due Date": next_due_date.strftime(DATE_FMT),
                    "Status": "Overdue" if next_due_date < today else "Due"
                })
        else:
             # Extinguisher never inspected
             due_inspections.append({
                "Serial Number": row['serial'],
                "Location": row['location'],
                "Last Inspection Date": "N/A",
                "Next Due Date": "Due Now",
                "Status": "Due"
            })

    due_df = pd.DataFrame(due_inspections)
    if not due_df.empty:
        st.dataframe(due_df)
    else:
        st.info("All active extinguishers are up-to-date with inspections.")


def generate_reports():
    st.subheader("Monthly Inspection Reports")

    extinguishers_df = get_extinguishers(active_only=False)
    inspections_df = get_inspections()

    if inspections_df.empty:
        st.info("No inspections recorded to generate reports.")
        return

    inspections_df['inspection_date'] = pd.to_datetime(inspections_df['inspection_date'])
    inspections_df['month_year'] = inspections_df['inspection_date'].dt.to_period('M')

    # Merge with extinguishers info
    report_df = pd.merge(inspections_df, extinguishers_df, left_on='extinguisher_id', right_on='id', suffixes=('_inspection', '_extinguisher'))

    # Select and rename columns for the report
    report_df = report_df[[
        'month_year',
        'serial',
        'location',
        'type',
        'inspection_date',
        'inspector',
        'gauge_ok',
        'seal_ok',
        'pin_ok',
        'hose_nozzle_ok',
        'clear_access',
        'notes_inspection'
    ]]
    report_df = report_df.rename(columns={'notes_inspection': 'inspection_notes'})


    st.dataframe(report_df)

    # Export as CSV
    csv_data = report_df.to_csv(index=False).encode('utf-8')
    st.download_button(
        label="Download Report as CSV",
        data=csv_data,
        file_name="fire_extinguisher_inspection_report.csv",
        mime="text/csv",
    )


def app_settings():
    st.subheader("Settings")

    st.write("### Import Extinguishers from CSV")
    st.write("Upload a CSV file with columns: 'Serial Number', 'Location', 'Type', 'Capacity (kg)', 'Manufacturer', 'Manufacture Date', 'In Service Date', 'Notes'")
    uploaded_file = st.file_uploader("Choose a CSV file", type="csv")

    if uploaded_file is not None:
        try:
            # Save the uploaded file temporarily
            filepath = os.path.join("/tmp", uploaded_file.name)
            with open(filepath, "wb") as f:
                f.write(uploaded_file.getbuffer())

            import_extinguishers_from_csv(filepath)
            st.success("Extinguishers imported successfully.")
            os.remove(filepath) # Clean up the temporary file
        except Exception as e:
            st.error(f"Error importing CSV: {e}")

if __name__ == "__main__":
    main()
```
